```python
# Step 2: Import required libraries
import os
import PyPDF2
import spacy
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from google.colab import files
```

```python
# Step 3: Define function to extract text from PDF
def extract_text_from_pdf(pdf_path):
    text = ""
    try:
        with open(pdf_path, 'rb') as file:
            reader = PyPDF2.PdfReader(file)
            for page_num in range(len(reader.pages)):
                text += reader.pages[page_num].extract_text()
    except Exception as e:
        print(f"Error extracting text from {pdf_path}: {e}")
    return text
```

```python
# Step 4: Define function to calculate cosine similarity
def calculate_similarity(text1, text2):
    nlp = spacy.load("en_core_web_sm")
    doc1 = nlp(text1)
    doc2 = nlp(text2)
    return doc1.similarity(doc2)
```

```python
# Step 3: Upload resume PDFs
uploaded = files.upload()  # Upload multiple resumes in PDF format
```

```python
# Step 4: Extract text from PDFs
def extract_text_from_pdf(file):
    reader = PyPDF2.PdfReader(file)
    text = ''
    for page in reader.pages:
        text += page.extract_text()
    return text

resume_texts = {}
for filename in uploaded.keys():
    text = extract_text_from_pdf(filename)
    resume_texts[filename] = text
```

```python
# Step 5: Load spaCy and preprocess text
nlp = spacy.load('en_core_web_sm')

def preprocess(text):
    doc = nlp(text.lower())
    tokens = [token.lemma_ for token in doc if token.is_alpha and not token.is_stop]
    return ' '.join(tokens)

processed_resumes = {name: preprocess(text) for name, text in resume_texts.items()}
```

```python
# Step 6: Provide a Full Stack Developer job description
job_description = """
We are looking for a Full Stack Developer proficient in front-end and back-end technologies.
Key skills include HTML, CSS, JavaScript, React, Node.js, Express, MongoDB, REST APIs, and Git.
Strong problem-solving and debugging skills required.
"""

processed_jd = preprocess(job_description)

```

```python
# Step 7: TF-IDF Vectorization and Scoring
docs = [processed_jd] + list(processed_resumes.values())
names = list(processed_resumes.keys())

vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(docs)

# Compute cosine similarity between JD and each resume
similarity_scores = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:]).flatten()
ranking = sorted(zip(names, similarity_scores), key=lambda x: x[1], reverse=True)
```

```python
# Step 8 (Updated): Display rankings with scores as whole numbers out of 100
df_ranked = pd.DataFrame(ranking, columns=["Resume", "Similarity"])
df_ranked["Score (/100)"] = (df_ranked["Similarity"] * 100).round(0).astype(int)  # Round and convert to int
df_ranked = df_ranked.drop(columns=["Similarity"])
df_ranked.index += 1
df_ranked

```

```python
# Step 9 (Updated): Export percentage scores as CSV
df_ranked.to_csv("ranked_resumes_percent.csv", index_label="Rank")
files.download("ranked_resumes_percent.csv")

```
